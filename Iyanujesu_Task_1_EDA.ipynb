```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Paths to the datasets
data_files = {
    "nigeria": r"C:\Users\Moses Akinyefa\Outreachy_Iyanujesu_2023\data\NGA_AL2_Nigeria.json",
    "nigeria_wsr": r"C:\Users\Moses Akinyefa\Outreachy_Iyanujesu_2023\data\NGA_AL2_Nigeria_WSR.json",
    "nigeria_gez": r"C:\Users\Moses Akinyefa\Outreachy_Iyanujesu_2023\data\NGA_AL2_Nigeria_GEZ.json",
    "lagos": r"C:\Users\Moses Akinyefa\Outreachy_Iyanujesu_2023\data\NGA_AL4_Lagos.json",
    "lagos_wsr": r"C:\Users\Moses Akinyefa\Outreachy_Iyanujesu_2023\data\NGA_AL4_Lagos_WSR.json",
    "lagos_gez": r"C:\Users\Moses Akinyefa\Outreachy_Iyanujesu_2023\data\NGA_AL4_Lagos_GEZ.json"
}

# Load the datasets into GeoDataFrames
gdfs = {name: gpd.read_file(path) for name, path in data_files.items()}
```

## Nationwide Analysis (Nigeria)


### Administrative Boundaries of Nigeria


First, let's visualize the administrative boundaries of Nigeria to understand the geographical scope of our analysis.


```python

# Function to plot administrative boundaries of Nigeria
def plot_administrative_boundaries(gdf, title):
    # Create a figure and axis
    fig, ax = plt.subplots(1, 1, figsize=(10, 10))
    
    # Plot the GeoDataFrame
    gdf.plot(ax=ax, color='lightgreen', edgecolor='black', linewidth=0.5)

    # Remove x and y axis labels and ticks
    ax.set_yticklabels([])
    ax.set_xticklabels([])
    ax.set_xticks([])
    ax.set_yticks([])

    # Set title
    ax.set_title(title, fontsize=15)

    # Display the plot
    plt.show()

# Call the function to plot the administrative boundaries of Nigeria
plot_administrative_boundaries(gdfs['nigeria'], "Administrative Boundaries of Nigeria")

```

### Observations
The map above displays the administrative boundaries of Nigeria. This visualization serves as a base map upon which we can overlay additional data layers, such as soil types and ecological zones.

Next, let's analyze and visualize the distribution of different soil types across Nigeria. Understanding soil types is crucial for various applications, including agriculture, land management, and conservation planning, as different soil types have different characteristics and are suitable for different uses.

### Soil Type Distribution Across Nigeria


Next, we explore the different soil types present across Nigeria. This is crucial for understanding land fertility, potential agricultural productivity, and necessary soil conservation measures.


```python
def plot_soil_types_further_corrected(gdf, title):
    # Create a figure and axis
    fig, ax = plt.subplots(1, 1, figsize=(10, 10))
    
    # Plot the GeoDataFrame
    gdf.plot(ax=ax, column='IPCC', legend=True, 
             legend_kwds={
                 'title': "Soil Types by IPCC classification", 
                 'bbox_to_anchor': (1, 1),  # places the legend outside, to the right and top
                 'loc': 'upper left'
             })
    
    # Remove x and y axis labels and ticks
    ax.set_yticklabels([])
    ax.set_xticklabels([])
    ax.set_xticks([])
    ax.set_yticks([])

    # Set title
    ax.set_title(title, fontsize=15)
    
    # Adjust the figure's layout to accommodate the legend
    plt.tight_layout(rect=[0, 0, 0.85, 1])  # adjust the right parameter to suit the size of your legend
    
    # Display the plot
    plt.show()

    return fig

# Plot the soil types of Nigeria with further corrected legend parameters
soil_type_map_further_corrected = plot_soil_types_further_corrected(gdfs['nigeria_wsr'], "Soil Types Distribution in Nigeria")

```

### Insights:

* Diverse Soil Types: Nigeria has a variety of soil types, which is indicative of the diverse agro-ecological zones. This diversity can support a wide range of agricultural practices.
* Agricultural Planning: Knowledge of soil types is vital for agricultural planning. Certain crops will thrive better in specific soil types. For instance, yam and cassava often do well in loamy soils, which are prevalent in certain regions of Nigeria.
* Land Management: Understanding the distribution of soil types can aid in sustainable land management practices. It helps in planning erosion control, fertilizer application, and land conservation practices.


Next, let's analyze and visualize the ecological zones within Nigeria. This information is crucial for environmental planning, conservation efforts, and understanding the biodiversity within the country.

### Ecological Zones Within Nigeria


Now, let's investigate the various ecological zones within Nigeria. These zones are critical for biodiversity conservation efforts and understanding the potential impacts of climate change.


```python
# Function to plot ecological zones across Nigeria
def plot_ecological_zones(gdf, title):
    # Create a figure and axis
    fig, ax = plt.subplots(1, 1, figsize=(10, 10))
    
    # Plot the GeoDataFrame
    gdf.plot(ax=ax, column='gez_name', legend=True, 
             legend_kwds={'title': "Ecological Zones",
                         'bbox_to_anchor': (1, 1),  # places the legend outside, to the right and top
                         'loc': 'upper left'})
    
    # Remove x and y axis labels and ticks
    ax.set_yticklabels([])
    ax.set_xticklabels([])
    ax.set_xticks([])
    ax.set_yticks([])

    # Set title
    ax.set_title(title, fontsize=15)

    plt.show()
    return fig

# Plot the ecological zones of Nigeria
ecological_zones_map = plot_ecological_zones(gdfs['nigeria_gez'], "Ecological Zones in Nigeria")

```

### Insights:

* Biodiversity Conservation: The diverse ecological zones indicate a rich biodiversity. Conservation efforts should consider these distinct zones to protect the unique ecosystems and species they support.

* Climate Change Resilience: Different ecological zones may have varying levels of resilience to climate change. Regions such as the "Tropical rainforest" are crucial carbon sinks and must be prioritized in climate change mitigation efforts.

* Sustainable Development: Understanding these zones is essential for sustainable development planning. For instance, regions classified as "Tropical dry forest" might be more suitable for certain types of agriculture but may also be more vulnerable to desertification.

* Environmental Policy: This knowledge is crucial for creating informed environmental policies. For example, the "Tropical mountain system" areas might require different environmental management strategies compared to the "Tropical shrubland" regions.

### Overlapping Analysis of Soil Types and Ecological Zones


Finally, we'll perform an integrated analysis to understand the overlap between soil types and ecological zones. This is vital for comprehensive land management strategies.


```python
# Function to visually inspect the overlap between ecological zones and soil types
def plot_overlap(gdf1, gdf2, title, legend_kwds1, legend_kwds2):
    # Create a figure and axis
    fig, ax = plt.subplots(1, 1, figsize=(15, 15))
    
    # Plot the first GeoDataFrame (soil types)
    gdf1.plot(ax=ax, column='IPCC', legend=True, 
              legend_kwds=legend_kwds1)
    
    # Plot the second GeoDataFrame (ecological zones) with half transparency
    gdf2.plot(ax=ax, column='gez_name', alpha=0.5, legend=True, 
              legend_kwds=legend_kwds2)

    # Remove x and y axis labels and ticks
    ax.set_yticklabels([])
    ax.set_xticklabels([])
    ax.set_xticks([])
    ax.set_yticks([])

    # Set title
    ax.set_title(title, fontsize=15)

    return fig

# Plot the overlap between soil types and ecological zones
overlap_map = plot_overlap(
    gdfs['nigeria_wsr'], 
    gdfs['nigeria_gez'], 
    "Overlap between Soil Types and Ecological Zones in Nigeria",
    {'title': "Soil Types by IPCC classification"},
    {'title': "Ecological Zones"}
)
plt.show()

```

### Insights:

Complex Interplay: The map shows a complex interplay between soil types and ecological zones, indicating that land management strategies need to be nuanced and location-specific.

Agriculture and Conservation: Areas where fertile soil types overlap with sensitive ecological zones might be high-priority for both agriculture and conservation. Policies in these areas must balance agricultural productivity with ecological preservation.

Climate Resilience: Regions where less fertile soil types overlap with ecologically sensitive zones might be prioritized for reforestation or other ecological restoration efforts to enhance climate resilience.

Data-Driven Decision Making: This kind of integrated analysis underscores the importance of data-driven decision-making in land management. It highlights areas where there may be competing interests or needs, guiding more informed policy and planning.

Further Analysis Required: While this visual overlap provides initial insights, more precise analysis (like what was initially intended with the spatial join) would be necessary for concrete policy formulation. This would involve actual data on the extent of overlap, the specific areas involved, and other contextual factors.

### BIODIVERSITY HOTSPOTS

```python
# load the biodiversity hotspots data
biodiversity_gdf = gpd.read_file(r'C:\Users\Moses Akinyefa\Outreachy_Iyanujesu_2023\data\CI_BiodiversityHotspots.geojson')
country_gdf = gdfs['nigeria']
state_gdf = gdfs['lagos']

# Check if any of the hotspots are within Nigeria's and Lagos's bounds
nigeria_bounds = gdfs['nigeria'].total_bounds  # Get total bounds of Nigeria
lagos_bounds = gdfs['lagos'].total_bounds  # Get total bounds of Lagos

# Filter the biodiversity hotspots to those that intersect with Nigeria's and Lagos's bounds
hotspots_in_nigeria = biodiversity_gdf[biodiversity_gdf.intersects(gdfs['nigeria'].unary_union)]
hotspots_in_lagos = biodiversity_gdf[biodiversity_gdf.intersects(gdfs['lagos'].unary_union)]

(hotspots_in_nigeria, hotspots_in_lagos)
```

The biodiversity hotspot that intersects with both Nigeria and Lagos is the "Guinean Forests of West Africa."

Now, we'll visualize this hotspot in relation to the administrative boundaries of Nigeria and Lagos. We'll create side-by-side visualizations for a comparative view. Let's proceed with this visualization.

```python
# Re-define the function and re-plot the biodiversity hotspots for Nigeria and Lagos
def plot_biodiversity_hotspots(country_gdf, state_gdf, hotspots_gdf, country_name, state_name):
    # Create a subplot with 2 panels: one for the country and one for the state
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 20))

    # Plot the country with hotspots
    gdfs['nigeria'].boundary.plot(ax=ax1, linewidth=1, color='black')
    hotspots_gdf.plot(ax=ax1, color='green', alpha=0.4)
    ax1.set_title(f'Biodiversity Hotspots in {country_name}', fontsize=15)
    ax1.set_xticks([])
    ax1.set_yticks([])

    # Plot the state with hotspots
    gdfs['lagos'].boundary.plot(ax=ax2, linewidth=1, color='black')
    hotspots_gdf.plot(ax=ax2, color='green', alpha=0.4)
    ax2.set_title(f'Biodiversity Hotspots in {state_name}', fontsize=15)
    ax2.set_xticks([])
    ax2.set_yticks([])

    plt.tight_layout()
    plt.show()

# Re-plot the biodiversity hotspots for Nigeria and Lagos
plot_biodiversity_hotspots(country_gdf, state_gdf, hotspots_in_nigeria, "Nigeria", "Lagos")
```

The visualizations display the biodiversity hotspots in relation to the administrative boundaries of Nigeria and Lagos, side by side.

* On the left, we see the entire country of Nigeria, with the biodiversity hotspot "Guinean Forests of West Africa" highlighted in green. This visualization provides a macro view, indicating the regions in Nigeria that are part of this biodiversity hotspot.

* On the right, we focus on Lagos. The same biodiversity hotspot covers part of Lagos, indicating that this area is of significant ecological importance.

These visualizations can be particularly useful for environmental conservation efforts, land use planning, and ecological research, as they clearly demarcate areas of high biodiversity that may require protective measures.


### Forest Cover

```python
def plot_forest_cover(country_gdf, state_gdf, country_name, state_name):
    # Create a subplot with 2 panels: one for the country and one for the state
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 20))

    # Plot the country's forest zones
    gdfs['nigeria_gez'].plot(ax=ax1, column='gez_name', cmap='viridis')
    ax1.set_title(f'Forest Cover in {country_name}', fontsize=15)
    ax1.set_xticks([])
    ax1.set_yticks([])

    # Create a legend for the country plot
    legend_elements = [plt.Line2D([0], [0], color='w', markerfacecolor='green', marker='o', markersize=10, label='Tropical dry forest'),
                       plt.Line2D([0], [0], color='w', markerfacecolor='blue', marker='o', markersize=10, label='Tropical moist forest'),
                       plt.Line2D([0], [0], color='w', markerfacecolor='red', marker='o', markersize=10, label='Tropical mountain system'),
                       plt.Line2D([0], [0], color='w', markerfacecolor='purple', marker='o', markersize=10, label='Tropical rainforest')]
    ax1.legend(handles=legend_elements, loc='upper left')

    # Plot the state's forest zones
    gdfs['lagos_gez'].plot(ax=ax2, column='gez_name', cmap='viridis')
    ax2.set_title(f'Forest Cover in {state_name}', fontsize=15)
    ax2.set_xticks([])
    ax2.set_yticks([])

    # Since Lagos has only one type of forest, we create a single legend element
    legend_elements_state = [plt.Line2D([0], [0], color='w', markerfacecolor='purple', marker='o', markersize=10, label='Tropical rainforest')]
    ax2.legend(handles=legend_elements_state, loc='upper left')

    plt.tight_layout()
    plt.show()

# Plot the forest cover for Nigeria and Lagos
plot_forest_cover(country_gdf, state_gdf, "Nigeria", "Lagos")
```

The visualizations display the forest cover in Nigeria and Lagos, represented by different colors for each type of forest:

* Nigeria (left panel):

The various colors represent different types of forests, including "Tropical dry forest," "Tropical moist forest," "Tropical mountain system," and "Tropical rainforest."
* Lagos (right panel):

Lagos is predominantly covered by "Tropical rainforest," as indicated by the color.
These visualizations provide valuable insights into the distribution of various forest types within Nigeria and specifically within Lagos. They can serve as essential tools for environmental management, conservation efforts, and land use planning, highlighting areas that might require protective measures or restoration efforts.
